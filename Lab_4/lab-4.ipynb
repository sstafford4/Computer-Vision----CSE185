```python
from ultralytics import YOLO 
import random
import cv2
import numpy as np
import os
import matplotlib.pyplot as plt
import pandas as pd
from collections import Counter

model = YOLO("/kaggle/working/yolov8m-seg-spotit.pt")

image_directory = "/kaggle/input/spot-it-eval" # using the 427 evaluation images Ross gave us. 
image_paths = [os.path.join(image_directory, fname) for fname in os.listdir(image_directory) if fname.lower().endswith(('.jpg'))]

yolo_classes = list(model.names.values())

class_colors = {i: tuple(random.choices(range(256), k=3)) for i in range(len(yolo_classes))}

output_dir = "/kaggle/working/segmented_images"
os.makedirs(output_dir, exist_ok=True)
csv_rows = []

for image_path in image_paths:
    img = cv2.imread(image_path)

    results = model.predict(img, conf=0.5)[0] # confidence threshold at 0.5

    detected_classes = [] # list to store all classes detected by the model

    if results.masks is not None: 
        for mask, box in zip(results.masks.xy, results.boxes):
            cls_id = int(box.cls[0]) # getting the class number to store in the list
            detected_classes.append(cls_id)

            points = np.int32([mask]) # drawing the mask
            color = class_colors.get(cls_id, (0, 255, 0))
            cv2.fillPoly(img, points, color)

    class_counts = Counter(detected_classes) # Counter pretty much just makes it a dictionary
    duplicate_classes = [cls_id for cls_id, count in class_counts.items() if count >= 2] # iterating through the dictionary to find dupes / matches

    if len(duplicate_classes) == 1: # if theres a single dupe, then thats the prediction. 
        predicted_class = yolo_classes[duplicate_classes[0]] 
    elif len(duplicate_classes) > 1: # if there are more than 1 dupes , then just take the chance and make the first one the pred
        #predicted_class = yolo_classes[duplicate_classes[0]] 
        # changed this to differentiate between correct guesses and approximate guesses. 
        predicted_class = "Too many choices: " + yolo_classes[duplicate_classes[0]]
    else:
        predicted_class = "N/A" # if no dupes, then no prediction. 

    filename = os.path.basename(image_path)

    csv_rows.append({"ID": filename, "Class": predicted_class}) # saving all of the predictions in the csv file

csv_filename = "/kaggle/working/Stafford.csv" 
df = pd.DataFrame(csv_rows)
df.to_csv(csv_filename, index=False)

```
